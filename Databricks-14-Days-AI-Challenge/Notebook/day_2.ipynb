```python
events = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
    header=True,
    inferSchema=True
)

events.printSchema()
```

```python
events.select(
    "event_type",
    "product_id",
    "brand",
    "price"
).show(10, truncate=False)
```

```python
# Count products with price > 100
events.filter(events.price > 100).count()
```

```python
events.groupBy("event_type").count().show()
```

```python
top_brands = (
    events
    .groupBy("brand")
    .count()
    .orderBy("count", ascending=False)
    .limit(5)
)

top_brands.show()
```

```python
top_brands.write.mode("overwrite").csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/top_brands"
)
```
