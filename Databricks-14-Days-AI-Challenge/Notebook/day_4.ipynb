Delta Lake Introduction

```python
from pyspark.sql import functions as F

events = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
    header=True,
    inferSchema=True
)

events.count()
```

```python
events.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.ecommerce.events_delta")
```

```python
events.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.ecommerce.events_delta")
```

```python
try:
    wrong_schema = spark.createDataFrame(
        [("a", "b", "c")],
        ["x", "y", "z"]
    )

    wrong_schema.write.format("delta") \
        .mode("append") \
        .save("/Volumes/workspace/ecommerce/delta/events")
except Exception as e:
    print("Schema enforcement triggered ✅")
    print(e)
```
